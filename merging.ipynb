```python
# 1. find total revenue generated

regs.merge(course, on='course_id')['price'].sum()
```

```python
# 2. find month by month revenue

temp_df = pd.concat([nov,dec],keys=['nov','dec']).reset_index()

temp_df.merge(course, on='course_id').groupby('level_0')["price"].sum()
```

```python
# 3. Print the registration table
# cols-> name-> course-> price

regs.merge(student, on='student_id').merge(course, on='course_id')[['name','course_name','price']]
```

```python
# 4. Plot bar chart for revenue/course

temp_df = regs.merge(student, on='student_id').merge(course, on='course_id')[['name','course_name','price']]
temp_df.groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
# 6. find course that got no enrollment
# courses['course_id']
# regs['course_id']

no_enrollment_ids = np.setdiff1d(course['course_id'], regs['course_id'])

course[course['course_id'].isin(no_enrollment_ids)]
```

```python
# 7. find students who did not enroll into any courses

no_enrolling_students = np.setdiff1d(student['student_id'],regs['student_id'])

student[student['student_id'].isin(no_enrolling_students)]
```

```python
# 8. Print student name-> partner name for all enrolled students
# self join

student.merge(student, left_on='student_id', right_on='partner', how='inner').merge(regs, left_on='student_id_x', right_on='student_id').merge(course, on='course_id')[['name_x','name_y','course_name']]
```

```python
student.merge(student, left_on='student_id', right_on='partner', how='inner')[['name_x','name_y']]
```

```python
# 9. find top 3 students who did most number enrollments

regs.merge(student, on='student_id', how='left').groupby('name').count()['student_id'].sort_values(ascending=False).head(3)
```
